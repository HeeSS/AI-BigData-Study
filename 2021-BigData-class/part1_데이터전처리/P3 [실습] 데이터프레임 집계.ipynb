# P3. [실습] 데이터프레임 집계

```python
# 라이브러리와 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# titanic 데이터 불러오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/titanic_train.csv'
titanic = pd.read_csv(path) 

# 확인
titanic.head()
```

1) 남녀별 평균 나이를 조회하세요.

```python
titanic.groupby(by=['Sex'], as_index=False)[['Age']].mean()
```

2) 객실등급(Pclass)별 평균 나이를 조회하세요.

```python
titanic.groupby(by=['Pclass'], as_index=False)[['Age']].mean()
```

3) 남녀별 탑승자 수를 조회하세요.

```python
titanic.columns
titanic.groupby(by=['Sex'], as_index=False)[['PassengerId']].count()
```

4) 남녀별 생존자 수를 조회하세요.

```python
titanic.groupby(by=['Sex'], as_index=False)['Survived'].sum()
```

5) 남녀별 생존율을 조회하세요.

- 0이 사망 1이 생존이므로 평균값이 생존율이 됩니다.

```python
titanic.groupby(by=['Sex'], as_index=False)['Survived'].mean()
```

6) 객실등급로 남녀별 금액(Fare)의 평균을 조회하세요.

```python
titanic.groupby(by=['Pclass', 'Sex'], as_index=False)['Fare'].mean()
```

7) 위 결과를 titanic_fare 라는 데이터프레임으로 선언학세요.

```python
titanic_fare = titanic.groupby(by=['Pclass', 'Sex'], as_index=False)['Fare'].mean()
titanic_fare
```

8) titanic_fare에서 Pclass가 1인 데이터에 대해서만 Sex 별 Fare를 **막대그래프**로 시각화 하세요.

```python
temp = titanic_fare.loc[titanic_fare['Pclass']==1]

plt.bar(temp['Sex'], temp['Fare'])
plt.show()
```

9) titanic_fare에서 Pclass가 1인 데이터에 대해서만 Sex 별 Fare를 **가로 막대그래프**로 시각화 하세요.

```python
temp = titanic_fare.loc[titanic_fare['Pclass']==1]

plt.barh(temp['Sex'], temp['Fare'])
plt.show()
```

10) Embarked가 S 인 탑승자의 남녀별 금액(Fare)의 표준편차를 조회하세요.

```python
temp = titanic.loc[titanic['Embarked']=='S']
temp.groupby(by=['Sex'], as_index=False)['Fare'].std()
```
