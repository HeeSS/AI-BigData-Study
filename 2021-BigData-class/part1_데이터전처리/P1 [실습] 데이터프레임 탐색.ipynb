7) 기초 통계 정보에서 gre 열의 25%, 50%, 75% 값을 q1, q2, q3 변수로 선언하고 확인하세요.

```python
q1 = graduate.describe()['gre']['25%']
q2 = graduate.describe()['gre']['50%']
q3 = graduate.describe()['gre']['75%']
print(q1, q2, q3)
```

8) gre 열 합과 평균을 gre_sum, gre_mean 변수로 선언하고 확인하세요.

```python
gre_sum = graduate['gre'].sum()
gre_mean = graduate['gre'].mean()
print(gre_sum, gre_mean)
```

9) gre 값 분포를 Histogram으로 확인하세요.

- 구간(Bins)을 20개로 나누세요.
- 빈도(Frequency) 대신 밀도(Density)를 표시하세요.

```python
plt.figure(figsize=(10,5))
plt.hist(graduate['gre'], bins=20, density=True)
plt.show()
```
