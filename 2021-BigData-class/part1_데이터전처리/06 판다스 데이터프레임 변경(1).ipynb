# 5.데이터프레임 변경(1)

- 데이터 전처리를 위해 꼭 익혀두어야 할 내용입니다.
- 반복 학습과 실습을 통해서 익숙해져야 합니다.

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 파일 불러오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/tips(2).csv'
tips = pd.read_csv(path, sep=',')  

# 확인
tips.head()
```

## 5.1. 열 이름 변경

- 기존 데이터프레임의 열 이름을 적절히 변경해야 할 경우가 있습니다.
- 또는 집계 결과를 가진 열 이름을 이해하기 쉽게 변경해야 할 경우도 있습니다.

**1) 일부 열 이름 변경**

- **rename()** 메소드를 사용해 변경 전후의 열 이름을 딕셔너리 형태로 나열하는 방법으로 변경합니다.
- **inplace=True** 옵션을 설정해야 변경 사항이 실제 반영이 됩니다.
- 다음과 같이 열 이름을 변경합니다.
    - total_bill_amount → total_bill
    - male_female → gender
    - smoke_yes_no → smoker
    - week_name → week
    - dinner_lunch → time

```python
# 연습을 위해 백업
tips_bak = tips.copy()

# rename() 함수로 열 이름 변경
tips.rename(columns={'total_bill_amount': 'total_bill',
                     'male_female': 'gender',
                     'smoke_yes_no': 'smoker',
                     'week_name': 'week',
                     'dinner_lunch': 'time'}, inplace=True)

# 확인
tips.head()
```

**2) 모든 열 이름 변경**

- 모든 열 이름을 변경할 때는 **columns** 속성을 변경합니다.
- 변경이 필요없는 열은 기존 이름을 부여해 변경합니다. 


```python
# tips 복원
tips = tips_bak.copy()

# 확인
tips.head()
```

```python
# 모든 열 이름 변경
tips.columns = ['total_bill', 'tip', 'gender', 'smoker', 'week', 'time', 'size']

# 확인
tips.head()
```

## 5.2. 열 추가

- 새로운 열을 추가하여 기존 데이터에서 계산된 결괏값을 저장해야할 경우가 있습니다.
- total_bill에 tip을 더한 값을 갖는 final_amt 열을 추가합니다.

```python
# final_amt 열 추가
tips['final_amt'] = tips['total_bill'] + tips['tip'] 

# 확인
tips.head()
```

<img src='https://raw.githubusercontent.com/jangrae/img/master/practice_01.png' width=120 align="left"/>

[문1]  total_bill을 size로 나눈 값을 갖는 div_tb 열을 추가하세요.

```python
# div_tb 열 추가
tips['div_tb'] = tips['total_bill'] / tips['size'] 

# 확인
tips.head()
```

[문2]  위에서 만들어진 div_tb를 x축으로하고 tip을 y축으로 해서 산점도를 표시하세요.

```python
# 산점도
plt.scatter(tips['div_tb'], tips['tip'])
plt.show()
```

[문3] 범주형 값을 갖는 week 열이 어떤 값을 몇 개 갖고 있는지 확인하세요.

```python
# 범주값 확인
tips['week'].value_counts()
```

[문4] week가 Sat, Sun 이면 1, 나머지는 0 값을 갖는 holiday 열을 추가하세요.

```python
# 열 추가
tips['holiday'] = 0
tips.loc[tips['week'].isin(['Sat', 'Sun']), 'holiday'] = 1

# 확인
tips.head()
```

## 5.3. 열 삭제

- 뭔가를 삭제할 때는 **항상 조심x100** 해야 합니다.

**1) 열 하나 삭제**

- **drop()** 메소드를 사용해 열을 삭제합니다.
- axis=0: 행 삭제(기본 값) 
- axis=1: 열 삭제
- **inplace=True** 옵션을 지정해야 실제로 반영이 됩니다.


```python
# 열 하나 삭제
tips.drop('final_amt', axis=1, inplace=True)

# 확인
tips.head()
```

**2) 여러 열 삭제**

- 삭제할 열을 리스트 형태로 전달해 한 번에 여러 열을 제거할 수 있습니다.

```python
# 열 두 개 삭제
tips.drop(['div_tb','week'], axis=1, inplace=True)

# 확인
tips.head()
```

## 5.4. 값 변경

- 분석을 돕기 위해 기존 값을 변경할 경우가 있습니다.
- 변경하는 방법은 여러가지가 있으니 편한 방법을 사용하면 됩니다.
- 예를 들어 조건에 따라 1과 0로 변경할 때 논리값을 사용할 수 있습니다.

```python
# 남자인가? -> True/False
male = tips['gender'] == 'Male'

# True/False -> 1/0 --> male 열 추가
tips['male'] = male.astype(int)

# 확인
tips.head()
```

- 또는 다음과 같이 확장 구문을 사용할 수 있습니다.

```python
# 확장 구문 사용
male = [1 if x=='Male' else 0 for x in tips['gender']]
tips['male'] = male

# 확인
tips.head()
```

- 새로운 열 추가후 불필요한 기존 열은 제거합니다.

```python
# 기존 열 삭제
tips.drop('gender', axis=1, inplace=True)

# 확인
tips.head()
```

[문1]  smoker 열이 어떤 범주값을 몇 개씩 갖는지 확인하세요.

```python
# 범주값 확인
tips['smoker'].value_counts()
```

[문2] smoker 열이 Yes/No 대신 1/0을 갖도록 논리값을 활용해 변경하세요.

```python
# 값 변경(Yes/No -> 1/0)
smoker = tips['smoker'] == 'Yes'
tips['smoker'] = smoker.astype(int)

# 확인
tips['smoker'].value_counts()
```

[문3] male 열이 갖는 두 값 1과 0에 대하여 tip 평균을 집계해 temp 데이터프레임으로 선언하세요.

```python
# tip 평균
temp = tips.groupby(by='male', as_index=False)['tip'].mean()
temp.head()
```

[문4] temp 데이터프레임의 male 열을 x, tip 열을 y로 하여 막대(bar) 그래프를 표시하세요.

```python
# bar chart
plt.bar(temp['male'], temp['tip'])
#plt.xticks([0, 1])
plt.xticks([0, 1], ['female', 'male'])
plt.title('Female Male Tips', fontsize=22, pad=20)
plt.xlabel('Gender')
plt.ylabel('Tip')
plt.show()
```

## 5.5. 유익한 기능

- 값을 변경하거나 결괏값을 갖는 열을 추가하는 유익한 기능들을 익혀두면 좋습니다.

```python
# 파일 불러오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/tips(2).csv'
tips = pd.read_csv(path, sep=',')  

# 열 이름 변경
tips.columns = ['total_bill', 'tip', 'gender', 'smoker', 'week', 'time', 'size']

# 확인
tips.head()
```

**1) map() 메소드**

- **map()** 메소드를 사용하면 범주형 값을 다른 값으로 쉽게 변경 할 수 있습니다.
- 다음 구문은 Male, Female을 각각 숫자 1, 0으로 변경합니다.

```python
# Male -> 1, Female -> 0
tips['gender'] = tips['gender'].map({'Male': 1, 'Female': 0})

# 확인
tips.head()
```

<img src='https://raw.githubusercontent.com/jangrae/img/master/practice_01.png' width=120 align="left"/>

[문1] smoker 열의 값 Yes와 No를 각각 숫자 1과 0으로 변경하세요.

```python
# Yes -> 1, No -> 0
tips['smoker'] = tips['smoker'].map({'Yes':1,'No':0})

# 확인
tips.head()
```

[문2] week 열이 어떤 범주값을 몇 개씩 가지는지 확인하세요.

```python
# 범주값 확인
tips['week'].value_counts()
```

[문3] week 값이 sat, sun이면 1, Thur, Fri면 0을 갖는 holiday 열을 map 메소드를 사용해 추가하세요.

```python
# Sat, Sun -> 1 / Thur, Fri -> 0
tips['holiday'] = tips['week'].map({'Sat':1, 'Sun':1, 'Thur':0, 'Fri':0})

# 확인
tips.head()
```

[문4] week 열을 삭제하세요.

```python
# 열 삭제
tips.drop('week', axis=1, inplace=True)

# 확인
tips.head()
```

**2) cut() 함수**

- **크기**를 기준으로 범위를 나누어 등급을 지정하고 싶을 때 **cut() 함수**를 사용하면 좋습니다.
- 범위 개수를 지정하면 자동으로 크기를 기준으로 나눕니다.

```python
# 크기로 5등분해서 등급 구하기
tips['tip_grp'] = pd.cut(tips['tip'], 5, labels=list('edcba') )

# 확인
tips['tip_grp'].value_counts()
```

- 임의의 범위를 직접 지정할 수 있습니다.
- 우선 tip 열의 기초통계량을 살펴봅니다.

```python
tips['tip'].describe()
```


- 다음 구문은 아래 범위에 따른 값을 갖는 tip_grp 열을 추가합니다.
  - 2.0 이하 → 'a'
  - 2.0 초과 2.9 이하 → 'b'
  - 2.9 초과 3.5625 이하 → 'c'
  - 3.5625 초과 10.0 이하 → 'd' 
- 음의 무한대는 -np.inf, 양의 무한대는 np.inf로 지정합니다.

```python
# 등급 구하기
bin = [-np.inf, 2.0, 2.9, 3.5625, 10.0]
#bin = [-np.inf, 2.0, 2.9, 3.5625, np.inf]
label = ['a', 'b', 'c', 'd']
tips['tip_grp'] = pd.cut(tips['tip'], bins=bin, labels=label)

# 확인
tips['tip_grp'].value_counts()
```

```python
# 확인
tips.head()
```

[문1] 기초통계량에서 total_bill 열의 25%, 50%, 75% 값을 조회해 q1, q2, q3 변수를 선언하세요.

```python
# 사분위수 
q1 = tips.describe()['total_bill']['25%']
q2 = tips.describe()['total_bill']['50%']
q3 = tips.describe()['total_bill']['75%']

# 확인
print(q1, q2, q3)
```

[믄2] 아래 범위와 값을 참조해 'a' ~ 'd' 값을 갖는 bill_grp 열을 추가하세요.

   - q1 이하 → 'a'
   - q1 초과 q2 이하 → 'b'
   - q2 초과 q3 이하 → 'c'
   - q3 초과 → 'd'

```python
# 등급 구하기
bin = [-np.inf, q1, q2, q3, np.inf]
label = ['a', 'b', 'c', 'd']
tips['bill_grp'] = pd.cut(tips['total_bill'], bins=bin, labels=label)

# 확인
tips.head()
```

[문3] bill_grp 별 total_bill, tip의 평균을 집계하세요.

```python
# 평균
tips.groupby(by='bill_grp', as_index=False)[['total_bill', 'tip']].mean()
```

```python
# 선 그래프
plt.plot(tips['tip'])
plt.show()
```
