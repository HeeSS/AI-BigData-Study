## 3.3. 시각화해서 보기

- 시각화를 통해 데이터를 좀더 쉽게 이해할 수 있습니다.

**1) Histogram**

- Histogram으로 데이터 분포를 쉽게 확인할 수 있습니다.

```python
# 히스토그램
plt.figure(figsize=(10,5))
plt.hist(titanic['Age'])
plt.show() # 이걸 해줘야 완성된거다.
```

- bins 옵션으로 구간 개수를 지정합니다. (디폴트 10)

```python
# 구간 개수 지정
plt.figure(figsize=(10,5))
plt.hist(titanic['Age'], bins=20)
plt.show()
```

- density=True를 지정해 빈도(Frequency) 대신 밀도(Density)를 확인할 수 있습니다.


```python
# Density 표시 - 면적을 다 더하면 1
plt.figure(figsize=(10,5))
plt.hist(titanic['Age'], bins=20, density=True)
plt.show()
```

```python
# 밀도 계산 예
40 / titanic['Age'].count() / 4   # 0.014005602240896359
```

**2) Scatter**

- Scatter(산점도)로 열 간의 상관관계를 확인할 수 있습니다.

```python
# 산점도
plt.figure(figsize=(10,5))
plt.scatter(titanic['Fare'], titanic['Pclass'])
#plt.scatter(titanic['Fare'], titanic['Survived']) # Pclass와 Survived의 상관관계 확인
plt.show()
```

**3) Line**

- Line Graph로 값의 변화를 쉽게 확인할 수 있습니다.
- x축, y축 값이 필요하지만 x축 값을 생략해도 됩니다.

```python
# Line
plt.figure(figsize=(10,5))
plt.plot(stock['Close'])
plt.show()
```

- 일부 데이터만 대상으로 집중해서 보고자 하면 행 수를 제헌하면 됩니다.
- 다음과 같이 앞 부분 4주(약 20일, 5 * 4 = 20) 정도를 대상으로 시각화합니다.

```python
# 초기 Close 변화
plt.figure(figsize=(10,5))
plt.plot(stock['Close'][:20])
plt.show()
```

```python
# 초기 Close 변화
plt.figure(figsize=(10,5))
plt.plot(stock['Close'].head(20))
plt.show()
```

- 다음과 같이 뒷 부분 4주 정도를 시각화 합니다.

```python
# 최근 Close 변화
plt.figure(figsize=(10,5))
plt.plot(stock['Close'][-20:])
plt.show()
```

- y 축에 값 두 개를 지정해 두 값을 비교해 볼 수 있습니다.

```python
# 최근 Open, Close 변화
plt.figure(figsize=(10,5))
plt.plot(stock[['Open', 'Close']][-20:])
#plt.legend(['Open', 'Close']) # 범례 표시
plt.show()
```

- 다음과 같이 조건에 맞는 데이터만을 대상으로 시각화할 수 있습니다.
- xticks() 함수로 축의 값을 기울일 수 있습니다.

```python
# 2021년 7월  데이터 추출 
stock_2021_07 = stock.loc[stock['Date'].between('2021-07-01', '2021-07-31')]

# 시각화
plt.figure(figsize=(10,5))
plt.plot(stock_2021_07['Date'], stock_2021_07[['Open', 'Close']])
plt.xticks(rotation=70) # x축의 컬럼값을 기울여 준다.
plt.show()
```

## 3.4. 정렬해서 보기

- 인덱스를 기준으로 정렬하는 방법과 특정 열을 기준으로 정렬하는 방법이 있습니다.
- **sort_index()** 메소드로 **인덱스**를 기준으로 정렬합니다.
- **sort_values()** 메소드로 **특정 열**을 기준으로 정렬합니다.
- **ascending** 옵션을 설정해 오름차순, 내림차순을 설정할 수 있습니다.
    - ascending=True: 오름차순 정렬(기본값)
    - ascending=False: 내림차순 정렬

**1) 인덱스를 기준으로 정렬**

```python
# 데이터프레임 확인
titanic.head()
```

```python
# 인덱스를 기준으로 내림차순 정렬
titanic.sort_index(ascending=False)
```

**2) 열 값을 기준으로 정렬**

- 하나 또는 여러 열을 기준으로 정렬할 수 있습니다.

```python
# Fare 열의 값으로 정렬
titanic.sort_values(by='Fare', ascending=False).head(10)
#titanic.sort_values('Fare', ascending=False) # 똑같다.
```

```python
# Fare, Age 열의 값으로 정렬
# fare를 내림차순으로 정렬하고, age를 오름차순으로 정렬
titanic.sort_values(by=['Fare', 'Age'], ascending=[False, True]).head()
```

## 3.5. 기본 집계

- 데이터를 좀더 이해하기 위해 고유값, 합, 평균, 최댓값, 최솟값 등을 확인합니다.

### 3.5.1. 고유값 확인

- 범주형 열(열이 가진 값이 일정한 값인 경우, 성별, 등급 등)인지 확인할 때 사용합니다.

**1) 고유값 확인**

- unique() 메소드로 고유값을 확인하며, 결괏값은 배열 형태가 됩니다.

```python
# Embarked 열 고유값 확인
titanic['Embarked'].unique()
```

**2) 고유값과 개수 확인**

- value_counts() 메소드로 고유값과 그 개수를 확인하며, 결괏값은 시리즈 형태가 됩니다.

```python
# Embarked 열 고유값 개수 확인, NaN 값 제외
titanic['Embarked'].value_counts()
```

```python
# Survived 열 고유값 개수 확인, NaN 값 제외
titanic['Survived'].value_counts()
```

```python
# sex 열 고유값 개수 확인, NaN 값 제외
titanic['Sex'].value_counts()
```

- dropna=False 옵션을 지정하면 NaN 값 개수도 확인할 수 있습니다.

```python
# Embarked 열 고유값 개수 확인, NaN값 포함
titanic['Embarked'].value_counts(dropna=False)
```

```python
# Pclass 열 고유값 개수 확인
titanic['Pclass'].value_counts(dropna=False)
```

### 3.5.2. 기본 통계 메소드 사용

- 데이터를 1차 집계 한 후 분석을 진행하는 경우가 많으므로 필히 알아두어야 할 내용입니다.
- 이후에 배우는 Groupby 기능에서 같이 사용됩니다.

```python
# 모든 열 합계 조회
titanic.sum()
```

```python
# 모든 열 평균값 조회
titanic.mean()
```

```python
# Fare 열 합계 조회
titanic['Fare'].sum()
```

```python
# Fare 열 최댓값 조회
titanic['Fare'].max()
```

```python
# Fare 열 표준편차 조회
titanic['Fare'].std()
```

```python
# Age, Fare 열 최솟값 조회
titanic[['Age', 'Fare']].min()
```

```python
# Age, Fare 열 평균값 확인
titanic[['Age', 'Fare']].mean()
```

```python
# Age, Fare 열 중앙값 조회
titanic[['Age', 'Fare']].median()
```
