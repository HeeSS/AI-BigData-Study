**3) print() 함수**

```python
# 단순 값 출력
print(10)

# 변수값 출력
a = 10
print(a)

# 연산 결과 출력
print(a + 100)
```

```python
# 구분자 지정
print('Hello', 'world.')
print('Hello', 'world.', sep=',')
print('Hello', 'world.', sep='')
```

```python
# 개행 여부 지정
print('Hello', 'world.')
print('Hello', 'world.')
print('Hello', 'world.', end='')
print('Hello', 'world.')
```

## 2.2. format 사용

- 문자열 일부분만 바꾸면서 해당 문자열을 재사용할 경우가 있습니다.
- 예를 들어 문자열에서 이름만 변경하며 사용하는 경우입니다.
- 이를 문자열 포맷팅이라고 하며, **문자열.format()** 형태로 사용합니다.

**1) 기본 사용법**

- 단순 문자열에서 { }는 아무 의미가 없습니다.

```python
# 단순 문자열
'이름:{}, 과목:{}, 점수:{}'
```

- 하지만 format을 만나면 { }는 큰 의미를 가집니다.

```python
# format 사용
'이름:{}, 과목:{}, 점수:{}'.format('홍길동', '국어', 90)
```

- format은 대부분 print() 함수 안에 사용합니다.

```python
# 변수 선언
name = '홍길동'
year = 2019
perf = 2345.6789

# 출력
print('이름:{}, 연도:{}, 실적:{}'.format(name, year, perf))
```

**2) 표시 위치 지정**

- { } 안에 변수 위치 값을 지정해야 하지만, 순서대로 사용될 것이면 생략하는 경우가 많습니다.
- { } 는 출력되는 값 순서 외에도 출력되는 형태를 제어할 수 있는 기능도 있습니다. 

```python
# 변수 선언
name = '홍길동'
year = 2019
perf = 2345.6789

# 출력
print('이름:{0}, 연도:{1}, 실적:{2}'.format(name, year, perf))
print('이름:{0}, 연도:{1}, 실적:{2:.2f}'.format(name, year, perf))
print('이름:{}, 연도:{}, 실적:{:,.2f}'.format(name, year, perf))
```

**4) upper(), lower(), title(), capitalize() 메소드**

**1) 기본 사용법**

- **range(*n*)**: 0 부터 n-1 까지의 정수 범위가 됩니다.
- **range(*m*, *n*)**: m 부터 n-1 까지의 정수 범위가 됩니다.

```python
# 0 ~ 9 정수를 요소로 갖는 리스트
nums = list(range(10))

# 확인
nums
```

```python
# -5 ~ 5 사이 정수를 요소로 갖는 리스트
nums = list(range(-5, 6))

# 확인
nums
```

**2) 증가값 지정**

- **range(*m, n, x*)**: m 부터 n-1 까지 x 만큼씩 증가하는 정수가 됩니다.

```python
# 0 ~ 9 사이, 2씩 증가하는 정수를 요소로 갖는 리스트
nums = list(range(0, 10, 2))

# 확인
nums
```

```python
# 0 ~ 9 사이, 3씩 증가하는 정수를 요소로 갖는 리스트
nums = list(range(0, 10, 3))

# 확인
nums
```

```python
# 9 ~ 0 사이, 1씩 감소하는 정수를 요소로 갖는 리스트
nums = list(range(9, -1, -1))

# 확인
nums
```

## 3.3. 리스트 인덱싱(Indexing)

- 특정 요소, 즉 리스트 안에서 특정 값을 찾을 때 사용하는 데이터를 **인덱스**라고합니다.

# 4.튜플(Tuple) 자료형

- 튜플은 **리스트와 정말 비슷한 자료형**입니다.
- 튜플은 대괄호([ ])가 아닌 **소괄호(( ))**를 사용해서 표현합니다.
- 튜플은 한 번 만들면 **요소의 값을 바꿀 수 없음**이 리스트와 다릅니다.
- 그 외는 리스트와 같은 특성을 가지므로 이해하기 쉬울 것입니다.

## 4.1. 튜플 만들기

- 리스트는 대괄호[[ ]]이고, 튜플은 소괄호(( ))인 사실만 기억하면 됩니다.
- 단, **괄호를 생략할 수 있다**는 점은 리스트와 다릅니다.
- 리스트에서는 대괄호 생략할 수 없습니다.

```python
# 여러 형태의 튜플 만들기
score1 = (90, 85, 70)
score2 = 90, 85, 70
score3 = (90, 85, 70, ('A', 'B', 'D', 'F'))

# 확인
print(score1)
print(score2)
print(score3)
```

```python
# range() 함수를 사용해 만들기
nums = tuple(range(10, 21,2))

# 확인
nums
```

## 4.2. 튜플 활용

- **괄호를 생략**할 수 있는 튜플의 특징을 이용하면 편리한 부분이 있습니다.

```python
# 괄호 없이 튜플 만들기
a = 10
b = 20
c = a, b
c
```

- 여러 변수에 동시에 값을 대입할 수 있습니다.

```python
# 여러 변수에 동시에 값 대입
a, b, c = 10, 20, 30
print('a =', a)
print('b =', b)
print('c =', c)
```

```python
# 여러 변수에 동시에 값 대입(튜플 포함)
a, b, c = 10, 20, (1, 2, 3)
print('a =', a)
print('b =', b)
print('c =', c)
```

**1) 합집합**

- **|** 연산자로 합집합을 구합니다.

```python
# 집합 만들기
member_set1 = {'홍길동', '한사랑', '일지매', '박여인'}
member_set2 = {'한사랑', '홍길민', '강우동'}

# 합집합 구하기
member_set3 = member_set1 | member_set2

# 확인
member_set3
```

```python
# 집합 만들기 
member_set1 = {'홍길동', '한사랑', '일지매', '박여인'}
member_set2 = {'한사랑', '홍길민', '강우동'}

# 교집합 구하가
member_set3 = member_set1 & member_set2

# 확인
member_set3
```

**3) 차집합**

- **-** 연산자로 차집합을 구합니다.

```python
# 집합 만들기
member_set1 = {'홍길동', '한사랑', '일지매', '박여인'}
member_set2 = {'한사랑', '홍길민', '강우동'}

# 차집합 구하기
member_set3 = member_set1 - member_set2

# 확인
member_set3
```

**4) 대칭 차집합**

- **^** 연산자로 대칭 차집합을 구합니다.

```python
# 집합 만들기
member_set1 = {'홍길동',  '한사랑', '일지매', '박여인'}
member_set2 = {'한사랑', '홍길민', '강우동'}

# 대칭 차집합 구하기
member_set3 = member_set1 ^ member_set2

# 확인
member_set3
```

**3) 반복문 확장**

- 확장 구문을 사용하면 반복문 코드가 매우 간결해집니다.
- 처음에는 다소 낯설 수 있으나, 이것이 **Python 다운 문법**이라 할 수 있습니다.
- 리스트 요소에 2를 곱한 결과를 갖는 새로운 리스트를 만드는 기본 구문은 다음과 같습니다.

```python
# 리스트 만들기
nums = [0, 1, 2, 3, 4]

# 빈 리스트 만들기
squares = []

# 2를 곱한 결과를 squares 리스트에 추가
for x in nums:
    squares.append(x ** 2)

# 확인
squares
```

- 위 구문을 확장 구문으로 작성하면 훨씬 간결해집니다.

```python
# 반복문 확장 문법 #1
squares = [x ** 2 for x in nums]

# 확인
squares
```

- 다음 구문과 같이 조건 처리도 가능합니다.

```python
# 반복문 확장 문법 #2 
even_squares = [x ** 2 for x in nums if x % 2 == 0] 

# 확인
even_squares
```

- 다음 구문의 의미를 잘 살펴보시기 바랍니다.

```python
# 리스트 선언
score = [70, 80, 75, 80, 90, 85, 95, 100, 90, 80]

# 요소별 개수 구하기
d_score = {x: score.count(x) for x in score}
print('빈도:', d_score)

# 최빈값 구하기
max_score = [x for x, y in d_score.items() if y == max(d_score.values())]
print('최빈값:', max_score)
```

# 8.함수(Function)

- 함수는 입력을 받아 무엇인가를 처리한 후 그 결과를 반환하는 것입니다.
- 쌀(입력)을 밥솥(함수)에 넣으면 밥(출력)이 되는 형태와 같습니다.
- 같은 입력(쌀)이라 해도 함수(밥솥)기능에 따라 다른 출력(밥)이 됩니다.

## 8.1. 함수 구조

```python
# 두 매개변수의 합을 반환하는 함수 만들기
def hap(x, y):
    return x + y

# 함수 사용
a = 10
b = 20
c = hap(a, b)
print("{0}와 {1}의 합은 {2}".format(a, b, c))
print("{0}와 {1}의 합은 {2}".format(10, 20, hap(10, 20)))
```

```python
# 매개변수가 음수인지, 양수인지 아니면 0 인지 확인하는 함수
def sign(x):
    if x > 0:
        return 'Positive'
    elif x < 0:
        return 'Negative'
    else:
        return 'Zero'

# 함수 사용
for x in [-15, 0, 12]:
    print('{:3d}: {}'.format(x, sign(x)))
```

### 8.2.1. 여러 유형의 함수

- 함수가 항상 매개변수와 리턴값을 갖는 것은 아닙니다.

**1) 매개변수와 리턴값이 있는 함수**

```python
# 두 매개변수의 합을 반환하는 함수
def hap(x, y):
    r = x + y
    return(r)

a = hap(10, 20)
a
```

**3) 매개변수는 있고 리턴값이 없는 함수**

### 8.2.2. 여러 매개변수가 있는 함수

- 일반적으로 여러 개의 매개변수가 있는 경우 함수 안에서 반복문을 사용해 처리합니다.

```python
# 임의의 여러 매개변수가 있는 함수
def hello(*args):
    for who in args:
        print('{}님 안녕하세요!'.format(who))

hello('홍길동', '일지매', '강우동')
```

```python
# 임의의 여러 매개변수와 고정된 매개변수가 있는 함수
def hello(*args, title):
    for who in args:
        print('{0}{1} 안녕하세요!'.format(who, title))

hello('홍길동', '일지매', '강우동', title='선생님')
```

### 8.2.3. 리턴값이 두 개인 함수

- 리턴값이 여러 개인 경우 튜플 형태로 반환됩니다.

**1) 리턴값이 두 개**

```python
# 리턴값이 두 개인 함수
def hap2(x, y):
    return x + y, x * y

r = hap2(10, 20)                 
r
```

**2) 리턴값이 세 개**

```python
# 리턴값이 세 개인 함수
def hap2(x, y):
    return x + y, x - y, x * y

r = hap2(10, 20)                 
r
```

**3) 여러 변수에 리턴값 대입**

- 이럴 때 튜플이 능력을 발휘합니다.

```python
# 리턴을 두 변수에 각각 받는 방법
def hap2(x, y):
    return x + y, x * y

s, m = hap2(10, 20)
print(s, m)
```

### 8.2.4. 매개변수 기본값을 갖는 함수

- 매개변수가 기본값을 가지면 함수를 호출할 때 매개변수를 전달하지 않아도 됩니다.
- 맨 뒷쪽 매개변수부터 기본값을 가질 수 있습니다.

```python
# 매개변수 기본값을 갖는 함수
def hello(who, title, end='!'):
    print('{0}{1} 안녕하세요{2}'.format(who, title, end))

hello('홍길동', '선생님', '?')
hello('홍길동', '선생님')
```

```python
# 중간의 매개변수는 기본값을 설정할 수 없음
def hello(who, end='!', title):
    print('{0}{1} 안녕하세요{2}'.format(who, title, end))

hello('홍길동', '선생님')
```

## 8.3. 함수와 변수

- 함수 안에서는 밖에서 선언한 변수를 변경할 수 없습니다.

```python
# 함수 안에서 밖의 변수를 변경할 수 없음
s1 = '홍길동'

def hello(s1):
    s1 = s1 + '님 안녕하세요!'

hello(s1)
s1
```

- 함수가 반환하는 값을 함수 밖의 변수가 받는 방법이 있습니다.

```python
# 반환되는 값을 받아 사용하기
s1 = '홍길동'

def hello(s1):
    return s1 + '님 안녕하세요!'

s1 = hello(s1)
s1
```

- global을 사용하면 함수 밖의 변수를 변경할 수 있습니다.

```python
# global 사용
s1 = '홍길동'

def hello():
    global s1
    s1 = s1 + '님 안녕하세요!'

hello()
s1
```
