## 2.1. 데이터프레임 이해

## 2.2. 데이터프레임 직접 만들기

- **pd.DataFrame()** 함수를 사용해 데이터프레임을 만듭니다.
- 데이터프레임을 만들기 위해서는 세 가지가 필요합니다.
    - 데이터
    - 열 이름(column)
    - 인덱스
- **데이터만 있어도** 우선은 데이터프레임을 **만들 수 있습니다.**
- 데이터프레임을 사용하기 위해 **Pandas** 라이브러리를 **pd** 별칭을 주어 불러옵니다.

```python
# 라이브러리 불러오기
import pandas as pd
```

**1) 데이터만 가지고 만들기**

- 인덱스와 열을 지정하지 않으면 행 번호가 인덱스, 열 번호가 열 이름이 됩니다.

```python
# 데이터
stock = [[94500, 92100, 92200, 92300],
         [96500, 93200, 95900, 94300],
         [93400, 91900, 93400, 92100],
         [94200, 92100, 94100, 92400],
         [94500, 92500, 94300, 92600]]

# 데이터프레임 만들기
df = pd.DataFrame(stock)

# 확인
df.head()
```

**2) 열 이름 지정**

- 열 이름이 될 데이터가 있으면 **columns** 옵션에 지정합니다.
- 당연히 열 이름은 열 개수만큼 지정해야 합니다.


```python
# 데이터
stock = [[94500, 92100, 92200, 92300],
         [96500, 93200, 95900, 94300],
         [93400, 91900, 93400, 92100],
         [94200, 92100, 94100, 92400],
         [94500, 92500, 94300, 92600]]

# 열 이름
names = ['High', 'Low', 'Open', 'Close']

# 데이터프레임 만들기
df = pd.DataFrame(stock, columns=names)

# 확인
df.head()
```

```python
# 열 이름 변경
df.columns = list('ABCD') # ['A','B','C','D']로 쪼개진다.
df.head()
```

**3) 인덱스  지정**

- 인덱스가 될 데이터가 있으면 **index** 옵션에 지정합니다.

```python
# 데이터
stock = [[94500, 92100, 92200, 92300],
         [96500, 93200, 95900, 94300],
         [93400, 91900, 93400, 92100],
         [94200, 92100, 94100, 92400],
         [94500, 92500, 94300, 92600]]

# 열 이름
names = ['High', 'Low', 'Open', 'Close']

# 인덱스
dates = ['2019-02-15', '2019-02-16', '2019-02-17', '2019-02-18', '2019-02-19']

# 데이터프레임 만들기
df = pd.DataFrame(stock, index=dates, columns=names)

# 확인
df.head()
```

## 2.3. CSV파일 읽어오기

- 분석용 데이터는 대부분 파일에서 읽어오니 잘 익혀야 할 기능입니다.
- **read_csv()** 함수를 사용해서 CSV 파일에서 데이터를 읽어옵니다.

**[주요옵션]**
- **sep**: 구분자 지정(기본값 = 콤마) 
- header: 헤더가 될 행 번호 지정(기본값 = 0)
- index_col: 인덱스 열 지정(기본값 = False)
- names: 열 이름으로 사용할 문자열 리스트
- skiprows: 처음 몇 줄을 무시할 것인지 지정, 리스트로 지정 가능
- **na_values**: NaN 값으로 간주할 값 
- **nrows**: 읽어올 행 수 

```python
# Titanic 데이터 불러오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/titanic_train.csv'
titanic = pd.read_csv(path)  

# 상위 10행만 확인
titanic.head(10)
```

- 특정 값을 NaN 값으로 간주해서 읽어와야 할 경우가 있습니다.

```python
# Titanic 데이터 불러오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/titanic_train(2).csv'
titanic = pd.read_csv(path)  

# 상위 10행만 확인
titanic.head(10)
```

```python
# Titanic 데이터 불러오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/titanic_train(2).csv'
titanic = pd.read_csv(path, na_values='-', sep=',') # '-'를 NaN으로 변환

# 상위 10행만 확인
titanic.head(10)
```

- 구분자가 꼭 콤마가 되라는 법은 없습니다.

```python
# Titanic 데이터 불러오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/titanic_train(3).csv'
titanic = pd.read_csv(path, sep='|')  

# 상위 10행만 확인
titanic.head(10)
```
