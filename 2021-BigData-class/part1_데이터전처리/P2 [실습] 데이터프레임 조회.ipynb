# P2. [실습] 데이터프레임 조회

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

- load_iris 모듈을 사용해 iris 데이터를 가져옵니다.

```python
# 모듈 불러오기
from sklearn.datasets import load_iris

# 데이터 가져오기
iris_src = load_iris()

# 새로운 변수로 선언
x = iris_src.data  
y = iris_src.target
x_col = iris_src.feature_names
```

1) 데이터프레임 값이 될 x, y, x_col 값을 확인하세요.

- 단, x와 y는 10개씩만 확인하세요.

```python
print(x[:10])
print(y[:10])
print(x_col)
```

2) x 값과 x_col을 사용해 iris 데이터프레임을 만드세요.

```python
iris = pd.DataFrame(x, columns = x_col)
iris.head()
```

3) 사용하기 쉽게 아래 지시된 열 이름으로 데이터프레임을 다시 만드세요.

- sepal_length, sepal_width, petal_length, petal_width

```python
x_col_name = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris = pd.DataFrame(x, columns = x_col_name)
iris.head()
```

4) y 값을 갖는 species 열을 iris 데이터프레임에 추가하세요.

```python
iris['species'] = y
iris.head()
```

- 다음 구문을 수행해서 species 열 값을 품종 이름으로 변경합니다.

```python
# 값 변경
iris.loc[iris['species']==0, 'species'] = 'setosa'
iris.loc[iris['species']==1, 'species'] = 'versicolor'
iris.loc[iris['species']==2, 'species'] = 'virginica'

# 확인
iris.head()
```

5) species 열에 어떤 값이 몇 개씩 있는지 확인하세요.

```python
iris['species'].value_counts()
```

6) sepal_length, sepal_width 열 평균을 각각 sl_mean, sw_mean 변수로 선언하고 확인하세요

```python
sl_mean = iris['sepal_length'].mean()
sw_mean = iris['sepal_width'].mean()
print(sl_mean, sw_mean)
```

7) sepal_length와 sepal_width 둘 다 평균 이상인 행을 조회하세요.

```python
iris.loc[(iris['sepal_length'] >= sl_mean) & (iris['sepal_width'] >= sw_mean)].head()
```

8) setosa 품종의 sepal_length 평균만 확인하세요.

```python
iris.loc[iris['species'] == 'setosa', 'sepal_length'].mean()
```

9) sepal_length와 species의 상관관계를 산점도로 시각화하세요.

```python
plt.figure(figsize=(10,5))
plt.scatter(iris['sepal_length'], iris['species'])
plt.show()
```

10) petal_length와 species의 상관관계를 산점도로 시각화하세요.

```python
plt.figure(figsize=(10,5))
plt.scatter(iris['petal_length'], iris['species'])
plt.show()
```
