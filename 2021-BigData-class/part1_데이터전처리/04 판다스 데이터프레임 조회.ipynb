## 4.3. 조건으로 조회

- **df.loc[조건]** 형태로 조건을 지정해 조건에 만족하는 데이터만 조회할 수 있습니다.
- 우선 조건이 제대로 판단이 되는지 확인한 후 그 **조건을 대 괄호 안에** 넣으면 됩니다.

```python
# graduate 데이터 불러오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/Graduate_apply.csv'
graduate = pd.read_csv(path, sep = ',')

# 확인
graduate.head()
```

**1) 단일 조건 조회**

```python
# gpa 열 값이 3.5 보다 큰 행 죠회
graduate.loc[graduate['gpa'] > 3.5]
```

- 조건을 변수로 선언해 사용할 수 있습니다.

```python
# 조건을 비교한 결과 True 인 값 조회
gpa_cond = graduate['gpa'] > 3.5
graduate.loc[gpa_cond]
```

**2) 여러 조건 조회**

- [ ]안에 조건을 여러개 연결할 때 **and와 or 대신에 &와 |**를 사용해야 합니다.
- 그리고 각 조건들은 **(조건1) & (조건2)** 형태로 **괄호**로 묶어야 합니다.

```python
# and로 여러 조건 연결
graduate.loc[(graduate['gpa'] > 3.0) & (graduate['rank'] == 3)]
```

```python
# or로 여러 조건 연결
graduate.loc[(graduate['gpa'] > 3.0) | (graduate['rank'] == 3)]
```

- 조건을 변수로 선언하면 코드를 단순화할 수 있습니다.

```python
# 조건을 변수로 선언
cond1 = graduate['gpa'] > 3.0
cond2 = graduate['rank'] == 3

# and 조건
graduate.loc[cond1 & cond2]
```

- 변수로 선언된 조건은 재사용할 때 사용 가치가 더 커집니다.

```python
# or 조건
graduate.loc[cond1 | cond2]
```

**3) 편리한 isin(), between() 메소드**

- **isin([값1, 값2,..., 값n])**: 값1 또는 값2 또는...값n인 데이터만 조회합니다.

```python
# 여러 값을 한꺼번에 나열해서 비교
graduate.loc[graduate['rank'].isin([1, 2])]
```

```python
# 위 문장을 분리하면
graduate.loc[(graduate['rank'] == 1) | (graduate['rank'] == 2)]
```

- **between(값1, 값2)**: 값1 ~ 값2까지 범위안의 데이터만 조회합니다.

```python
# 어디부터 어디까지 범위를 한 번에 지정
graduate.loc[graduate['rank'].between(1, 3)]
```

```python
# 위 문장을 분리하면
graduate.loc[(graduate['rank'] >= 1) & (graduate['rank'] <= 3)]
```

**4) 조건을 만족하는 행의 일부 열 조회**

- **df.loc[조건, ['열 이름1', '열 이름2',...]]** 형태로 조회할 열을 리스트로 지정합니다.

```python
# 조건에 맞는 하나의 열 조회
graduate.loc[graduate['gpa'] > 3.0, ['gre']]
```

```python
# 조건에 맞는 여러 열 조회
graduate.loc[graduate['gpa'] > 3.0, ['admit', 'gpa', 'gre']]
```

## 4.4. 인덱스 초기화

- 다음과 같은 경우 인덱스 초기화가 필요합니다.
    - 기존 데이터프레임에서 일부 행을 가져와 새로 만든 데이터프레임
    - 일부 행이 지워진 데이터프레임
- 특히 0부터 시작하는 정수형 인덱스의 경우 초기화를 하는 것이 좋습니다.
- 깔끔한 데이터가 사용하기 좋은 데이터입니다.

```python
# rank가 1인 행 추축
rank_1 = graduate.loc[graduate['rank'] == 1]

# 확인
rank_1
```

- **reset_index()** 메소드로 인덱스를 초기화합니다.
- 기존 인덱스 값은 의미가 없으므로 drop=True 옵션을 지정해 제거합니다.
- inplace=True 옵션을 지정하지 않으면 실제 반영되지 않습니다.


```python
# 인덱스 초기화
rank_1.reset_index(drop=True, inplace=True)
#rank_1 = rank_1.reset_index(drop=True) # 위 코드와 같은의미, inplace=True 하면 안된다.

# 확인
rank_1
```
